```python
#Importacao de bibliotecas
import pandas as pd
import numpy as np
```

```python
#Data frame dos casos
df_casos = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
df_casos.tail()
```

```python
#Encontrar o tamanho da matriz
df_casos.shape
```

```python
#Selecao da colunas numericas
df_casos.iloc[ : , 4: df_casos.shape[1]]
```

```python
#Soma dos casos
df_casos = np.sum(df_casos.iloc[ : , 4: df_casos.shape[1]])
df_casos.index = pd.DatetimeIndex(df_casos.index)
df_casos.tail()
```

```python
df_casos.plot()
```

## Mortes por COVID-19

```python
df_mortes = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
```

```python
df_mortes.shape
```

```python
#Soma de mortes Covid-19
df_mortes = np.sum(df_mortes.iloc[ : , 4: df_mortes.shape[1]])
df_mortes.index = pd.DatetimeIndex(df_mortes.index)
df_mortes.tail()
```

```python
df_casos.plot()
```

```python
#Mortalidade
mortalidade = (df_mortes/df_casos)*100
```

```python
mortalidade.plot()
```

```python
df_recuperados = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
```

```python
df_recuperados.shape
```

```python
#Soma dos recuperados

df_recuperados = np.sum(df_recuperados.iloc[ : , 4: df_recuperados.shape[1]])
df_recuperados.index = pd.DatetimeIndex(df_recuperados.index)
df_recuperados.tail()
```

```python
df_recuperados.plot()
```

```python
df_analise = pd.concat([df_casos, df_mortes, df_recuperados], axis=1)
df_analise.columns=(['casos', 'mortes', 'recuperados'])
```

```python
df_analise.tail()
```

```python
df_analise.plot()
```
